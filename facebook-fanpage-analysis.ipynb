```python
data = pd.read_csv('https://drive.google.com/uc?export=download&id=1DJlWsDV2m4wyRmleakkAUoS5Crvvgrgc',index_col=0)

print("原始資料尺寸:",data.shape)
#print("資料概觀",data.head)
data=data.dropna()
print("去除缺失資料後尺寸:",data.shape)
a=data.values
```

```python
c=0
for i in range(0,a.shape[0],1):
    if a[i-c,6]<100:
        #print(a[i-c,:],c)
        a=np.delete(a,i-c,0)
        c+=1
print("去除問題資料後尺寸::",a.shape)
```

```python
import matplotlib.pyplot as plt
for i in range(0,a.shape[0],1):
    if a[i-c,6]<100:
        print('fff')
print("去除問題資料後尺寸::",a.shape)
plt.figure(figsize=(16, 8), dpi=80)
plt.scatter(a[:,6],a[:,5]);
plt.show()
```

```python
b=np.array([a[:,4],a[:,6]])
c=b.T
print(b)
print(c)
```

```python

X = c
print(X.shape)
y = a[:,5]
print(y.shape)
lm = LinearRegression()
lm.fit(X, y)
print("印出係數",lm.coef_)
print("印出截距",lm.intercept_ )
```

```python
to_be_predicted = np.array([
    [76052, 2150]
])
predicted_sales = lm.predict(to_be_predicted)

# 預測第一筆資料
print(predicted_sales)
#110149
# 模型績效
mse = np.mean((lm.predict(X) - y) ** 2)
r_squared = lm.score(X, y)  #R 告訴你在你的模型裡，有多少的 variance 是可以被你的自變數解釋的。
adj_r_squared = r_squared - (1 - r_squared) * (X.shape[1] / (X.shape[0] - X.shape[1] - 1))

# 印出模型績效
print(mse)
print(r_squared)
print(adj_r_squared)
```

```python

plt.scatter(X[:,0], y, color='black')
plt.plot(X[:,0], lm.predict(X), color='blue', linewidth=3)
#plt.plot(to_be_predicted, predicted_sales, color = 'red', marker = '^', markersize = 10)
plt.xlabel("Lifetime Post organic reach")
plt.ylabel("Lifetime Post Organic Impressions")
#plt.xticks(())
#plt.yticks(())
plt.show()
```
